```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("insurance_claims_raw.xlsx")
```

```python
new_data=df.drop("_c39",axis=1)
```

```python
grouped_df = df.groupby(['incident_type', 'fraud_reported']).size().reset_index(name='counts')

# Pivot the data for better visualization
pivot_df = grouped_df.pivot(index='incident_type', columns='fraud_reported', values='counts').fillna(0)

# Plotting
pivot_df.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Incident Type and Fraud Reported')
plt.xlabel('Incident Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Fraud Reported')
plt.show()
```

# Feature Engineering

let us look out for missing values

```python
new_data.isnull().sum()
```

The below code is used to fill Null Values by imputing them with mean value for columns containing integers and impute with mode for columns containing string values

```python
new_data["age"]=df["age"].fillna(df["age"].mean())
new_data["policy_annual_premium"]=df["policy_annual_premium"].fillna(df["policy_annual_premium"].mean())
new_data["policy_deductable"]=df["policy_deductable"].fillna(df["policy_deductable"].mean())
new_data["authorities_contacted"]=df["authorities_contacted"].fillna(df["authorities_contacted"].mode()[0])
new_data["incident_state"]=df["incident_state"].fillna(df["incident_state"].mode()[0])                                                               
new_data["insured_hobbies"]=df["insured_hobbies"].fillna(df["insured_hobbies"].mode()[0])
new_data["capital-gains"]=df["capital-gains"].fillna(df["capital-gains"].mode()[0])
new_data["insured_education_level"]=df["insured_education_level"].fillna(df["insured_education_level"].mode()[0])
```

In order for Machine Learning Algorthms to build models it need to interact with columns that are numeric. As a result we need to use Sckit-learn Algorthm (LabelEncoder) to create dummies for each an every column that contist of categories which are strings

```python
label_encoder=LabelEncoder()
```

```python
new_data["insured_sex"]=label_encoder.fit_transform(df["insured_sex"])
new_data["fraud_reported"]=label_encoder.fit_transform(df["fraud_reported"])
new_data["police_report_available"]=label_encoder.fit_transform(df["police_report_available"])
new_data["property_damage"]=label_encoder.fit_transform(df["property_damage"])
new_data["incident_severity"]=label_encoder.fit_transform(df["incident_severity"])
new_data["collision_type"]=label_encoder.fit_transform(df["collision_type"])
```

```python
new_data["months_as_customer"]=new_data["months_as_customer"]/12
```

```python
def age_groupes(age):
    if age>=20 and age<30:
        return "young Adults"
    elif age<=30 and age<40:
        return "Adults"
    else:
        return "middle age adult"
```

```python
new_data["new_age"]=new_data["age"].apply(age_groupes)
```

```python
grouped_data=new_data.groupby("new_age")["fraud_reported"].sum().reset_index()
```

```python
grouped_data.plot(kind="bar",x="new_age",y="fraud_reported",stacked=True)
plt.xlabel('Age Range')
plt.ylabel('Number of Fraud Reports')
plt.title('Fraud Reports by Age Range')
plt.show()
```

```python
grouped_df2= new_data.groupby(['new_age', 'fraud_reported']).size().reset_index(name='counts')

# Pivot the data for better visualization
pivot_df2 = grouped_df2.pivot(index='new_age', columns='fraud_reported', values='counts').fillna(0)

# Plotting
pivot_df2.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Incident Type and Fraud Reported')
plt.xlabel('Incident Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Fraud Reported')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming new_data is your DataFrame

# Group by 'new_age' and 'fraud_reported' and get the counts
grouped_df2 = new_data.groupby(['new_age', 'fraud_reported']).size().reset_index(name='counts')

# Pivot the data for better visualization
pivot_df2 = grouped_df2.pivot(index='new_age', columns='fraud_reported', values='counts').fillna(0)

# Plotting pie charts for each age range
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))
axes = axes.flatten()

for i, (age, data) in enumerate(pivot_df2.iterrows()):
    data.plot(kind='pie', ax=axes[i], autopct='%1.1f%%', startangle=90)
    axes[i].set_title(f'Age Range: {age}')
    axes[i].set_ylabel('')

plt.suptitle('Fraud Reported by Age Range')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming new_data is your DataFrame

# Group by 'new_age' and 'fraud_reported' and get the counts
grouped_df2 = new_data.groupby(['new_age', 'fraud_reported']).size().reset_index(name='counts')

# Pivot the data for better visualization
pivot_df2 = grouped_df2.pivot(index='new_age', columns='fraud_reported', values='counts').fillna(0)

# Plotting donut charts for each age range
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))
axes = axes.flatten()

for i, (age, data) in enumerate(pivot_df2.iterrows()):
    data.plot(kind='pie', ax=axes[i], autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.3))
    axes[i].set_title(f'Age Range: {age}')
    axes[i].set_ylabel('')

plt.suptitle('Fraud Reported by Age Range')
plt.tight_layout()
plt.show()
```
