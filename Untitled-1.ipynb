# Feature Engineering

let us look out for missing values

```python
bin_edges = [0, 30, 55, 100]  # Define the bin edges
bin_labels = ['Young Adult', 'Middle Aged', 'Elderly']  # Corresponding labels for each bin

# Create a new column based on the bin labels
new_data['ages_category'] = pd.cut(new_data['age'], bins=bin_edges, labels=bin_labels)
```

```python
bin_edges_customer = [0, 25, 150, 500]  # Define the bin edges
bin_labels_customer = ['New Client', 'Established Client', 'Long-Term Client']  # Corresponding labels for each bin

# Create a new column based on the bin labels
new_data['customer_category'] = pd.cut(new_data['months_as_customer'], bins=bin_edges_customer, labels=bin_labels_customer)
```

```python
new_data['total_premiums_paid'] = (new_data['policy_annual_premium']/12) * new_data['months_as_customer']
```

```python
new_data['net_value_of_customer'] = new_data['total_premiums_paid'] - new_data['total_claim_amount']
```

```python
new_data["Contract Years"] = new_data["months_as_customer"]/12
```

The below code is used to fill Null Values by imputing them with mean value for columns containing integers and impute with mode for columns containing string values

```python
def replace_values_all_columns(new_data, to_replace, replacement=pd.NA):
    # Loop through all columns and replace the specified value
    df.replace(to_replace, replacement, inplace=True)
    return df

# Usage example
new_data = replace_values_all_columns(new_data, '?')
```

```python
def imputting_columns(new_data):
    for column in new_data.columns:
        if new_data[column].dtype=="object":
            new_data[column]=new_data[column].fillna(new_data[column].mode()[0])
        else:
            new_data[column]=new_data[column].fillna(new_data[column].mean())

    return new_data

new_data=imputting_columns(new_data)
```

```python
from sklearn.preprocessing import LabelEncoder
def dumming_columns(new_data):
    label_encoder=LabelEncoder()
    for column in new_data.columns:
        if new_data[column].dtype=="object":
            new_data[column] = new_data[column].astype(str)
            new_data[column]=label_encoder.fit_transform(new_data[column])
    return new_data

new_data=dumming_columns(new_data)
```
