This function automates the process of handling missing values in a DataFrame by imputing them with the most frequent value for categorical data and the mean for numerical data, ensuring that the DataFrame is ready for further analysis.

```python
def imputting_columns(new_data):
    for column in new_data.columns:
        if new_data[column].dtype=="object":
            new_data[column]=new_data[column].fillna(new_data[column].mode()[0])
        elif pd.api.types.is_numeric_dtype(new_data[column]):
            new_data[column]=new_data[column].fillna(new_data[column].mean())
            
    return new_data

new_data=imputting_columns(new_data)
```

Machine learning algorithms require numerical input to build models effectively. Therefore, any categorical data (i.e., columns containing strings) must be converted into numeric values. One common way to achieve this is by using the LabelEncoder from the scikit-learn library, which transforms categorical values into numeric labels.

```python
def encode_columns(df):
    label_encoder = LabelEncoder()

    # Loop through each column in the DataFrame
    for column in df.columns:
        # Check if the column is of object type or contains strings
        if df[column].dtype == "object" or df[column].apply(lambda x: isinstance(x, str)).any():
            # Apply LabelEncoder to the column
            df[column] = label_encoder.fit_transform(df[column].astype(str))

    return df
new_data_encoded = encode_columns(new_data)
```

# Train Test Split


Out-of-sample accuracy refers to the percentage of correct predictions made by a model on data that it has not been trained on. When a model is trained and tested on the same dataset, it often exhibits low out-of-sample accuracy due to overfitting.

Ensuring high out-of-sample accuracy is crucial because the ultimate goal of any model is to accurately predict outcomes on unseen data. One effective method to improve out-of-sample accuracy is through a technique called Train/Test Split.

This approach involves dividing the dataset into two mutually exclusive sets: a training set and a testing set. The model is trained on the training set and then evaluated on the testing set. By using data that the model hasn't seen during training, Train/Test Split provides a more realistic assessment of the model's performance and enhances its ability to generalize to unseen data in real-world scenarios.

```python
# Split the data into training and testing sets
X_train, X_val, y_train, y_val = train_test_split(x,y, test_size=0.2, random_state=42)
```

# Modelling

```python
def fit_and_validate_models(X_train, X_val, y_train, y_val):
    # Standardize the data for models that require it (SVM and Neural Networks)
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_val)
    
    # Initialize models
    models = {
        "Logistic Regression": LogisticRegression(max_iter=1000),
        "Random Forest": RandomForestClassifier(n_estimators=100),
        "Support Vector Machine": SVC(probability=True),  # For ROC AUC, probability=True is needed
        "Neural Network": MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000),
        "DecisionTreeClassifier":DecisionTreeClassifier(random_state=42),
         "AdaBoost": AdaBoostClassifier(n_estimators=100)
        
    }
    
    # Initialize performance metrics dictionary
    performance_metrics = {}
    
    # Fit and validate each model
    for model_name, model in models.items():
        # Use scaled data for SVM and Neural Network, original data for others
        if model_name in ["Support Vector Machine", "Neural Network"]:
            model.fit(X_train_scaled, y_train)
            y_pred = model.predict(X_test_scaled)
            y_prob = model.predict_proba(X_test_scaled)[:, 1]
        else:
            model.fit(X_train, y_train)
            y_pred = model.predict(X_val)
            y_prob = model.predict_proba(X_val)[:, 1]
        
        # Calculate performance metrics
        accuracy = accuracy_score(y_val, y_pred)
        precision = precision_score(y_val, y_pred)
        recall = recall_score(y_val, y_pred)
        f1 = f1_score(y_val, y_pred)
        roc_auc = roc_auc_score(y_val, y_prob)
        
        # Store performance metrics for the model
        performance_metrics[model_name] = {
            "Accuracy": accuracy,
            "Precision": precision,
            "Recall": recall,
            "F1 Score": f1,
            "ROC AUC Score": roc_auc
        }
    
    return performance_metrics

# Example usage:
performance = fit_and_validate_models(X_train, X_val, y_train, y_val)
performance
```

# Model explanation

Explanation of the Metrics:

**Accuracy**: The proportion of correct predictions over the total predictions. While accuracy is a useful metric, it can be misleading in imbalanced datasets, where the minority class (e.g., fraud) is underrepresented.

**Precision**: The proportion of true positive predictions over the total positive predictions. Precision is crucial when the cost of a false positive (e.g., wrongly identifying a legitimate claim as fraud) is high.

**Recall**: The proportion of true positives over the total actual positives. High recall is important when missing positive cases (e.g., actual fraudulent claims) has severe consequences.

**F1 Score**: The harmonic mean of precision and recall. The F1 Score balances the trade-off between precision and recall and is especially useful in cases of imbalanced classes, where one class is significantly more frequent than the other.

```python
def plot_performance_metrics(performance_metrics):
    # Convert performance metrics to DataFrame
    metrics_df = pd.DataFrame(performance_metrics).T
    
    # Create a figure with subplots (3 rows, 2 columns)
    fig, axes = plt.subplots(3, 2, figsize=(15, 15))
    fig.suptitle('Model Performance Metrics', fontsize=16)
    
    metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'ROC AUC Score']
    
    # Plot each metric
    for i, metric in enumerate(metrics):
        ax = axes[i // 2, i % 2]
        sns.barplot(x=metrics_df.index, y=metrics_df[metric], ax=ax, palette='viridis')
        ax.set_title(metric)
        ax.set_xlabel('Model')
        ax.set_ylabel(metric)
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

    # Hide unused subplots
    for j in range(len(metrics), 6):
        fig.delaxes(axes[j // 2, j % 2])
    
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

# Example usage:
plot_performance_metrics(performance)
```
